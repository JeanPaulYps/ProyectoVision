```python
# Importar paquetes necesarios

from mrcnn.config import Config
from mrcnn import model as modellib
from mrcnn import visualize
import numpy as np
import colorsys
import argparse
import imutils
import random
import cv2
import os
import matplotlib.pyplot as plt




# Cargar los  class label names del disco, one label por linea
CLASS_NAMES = open("coco_labels.txt").read().strip().split("\n")

# Generar un aleatorio, pero distintio color para cada class label

hsv = [(i / len(CLASS_NAMES), 1, 1.0) for i in range(len(CLASS_NAMES))]
COLORS = list(map(lambda c: colorsys.hsv_to_rgb(*c), hsv))
random.seed(42)
random.shuffle(COLORS)

class SimpleConfig(Config):
	# Dar un nombre a la configuracin general
	NAME = "coco_inference"

	# cargar el nmero de  GPUs a usar con el nmero de imagenes por GPU
	GPU_COUNT = 1
	IMAGES_PER_GPU = 1

# Numero de clases

	NUM_CLASSES = len(CLASS_NAMES)

# Inicializar la configuracin de inferencia
config = SimpleConfig()

# Inicializar el  Mask R-CNN modelo para inferenica y entonces cargar los pesos

print("[INFO] loading Mask R-CNN model...")
model = modellib.MaskRCNN(mode="inference", config=config,
	model_dir=os.getcwd())
model.load_weights("mask_rcnn_coco.h5", by_name=True)

# Cargar la imagen de entrada y convertir esta de  BGR a RGB
# ordenar y rescalar la imagen
image = cv2.imread("img/pasillo.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = imutils.resize(image, width=512)

# Realizar un forward pass de la red neuronal para obtener los resultados
print("[INFO] Haciendo predicciones con Mask R-CNN...")
r = model.detect([image], verbose=1)[0]

# loop over of the detected object's bounding boxes and masks
for i in range(0, r["rois"].shape[0]):
	# extract the class ID and mask for the current detection, then
	# grab the color to visualize the mask (in BGR format)
	classID = r["class_ids"][i]
	mask = r["masks"][:, :, i]
	color = COLORS[classID][::-1]

	# Visualizar la mascara en pixeles del objeto
    
	image = visualize.apply_mask(image, mask, color, alpha=0.5)

# Convertir la imagen de regreso a BGR para poder usar OpenCV's drawing functions

image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

# ciclo  sobre los  scores calculados and los class labels

for i in range(0, len(r["scores"])):
	# Extraer la informacin de la caja, class ID, label, prediccin
	# probabilidad y color de visualizacin
	(startY, startX, endY, endX) = r["rois"][i]
	classID = r["class_ids"][i]
	label = CLASS_NAMES[classID]
	score = r["scores"][i]
	color = [int(c) for c in np.array(COLORS[classID]) * 255]

	# dibujar cajas, class label, and score de el objeto
	cv2.rectangle(image, (startX, startY), (endX, endY), color, 2)
	text = "{}: {:.3f}".format(label, score)
	y = startY - 10 if startY - 10 > 10 else startY + 10
	cv2.putText(image, text, (startX, y), cv2.FONT_HERSHEY_SIMPLEX,
		0.6, color, 2)

# Mostrar la imagen resultado

plt.figure(figsize=(20,20))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
